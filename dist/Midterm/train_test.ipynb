### 補值，數值型用中位數；物件型用眾數補

```python
# 定義類別型態的欄位名稱
categorical_columns = [
    'txkey', 'chid', 'cano', 'contp', 'etymd', 'mchno', 'acqic', 'mcc', 'ecfg', 'insfg', 'bnsfg', 
    'stocn', 'scity', 'stscd', 'ovrlt', 'flbmk', 'hcefg', 'csmcu', 'flg_3dsmk'
]
```

```python
# 處理訓練資料集的缺失值
train_dataset[categorical_columns] = train_dataset[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))
train_dataset = train_dataset.apply(lambda x: x.fillna(x.median()) if x.name not in categorical_columns else x)

# 處理測試資料集的缺失值
test_X[categorical_columns] = test_X[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))
test_X = test_X.apply(lambda x: x.fillna(x.median()) if x.name not in categorical_columns else x)
```

```python
# 檢查訓練資料集都補了什麼值
for column in train_dataset.columns:
    if column in categorical_columns:
        print(f'{column}: {train_dataset[column].mode()[0]}')
    elif column == 'label':
        pass
    else:
        print(f'{column}: {train_dataset[column].median()}')
```

```python
# 檢查測試資料集都補了什麼值
for column in test_X.columns:
    if column in categorical_columns:
        print(f'{column}: {test_X[column].mode()[0]}')
    else:
        print(f'{column}: {test_X[column].median()}')
```

## 類別資料轉label encoding

```python
train_X = train_dataset.drop(columns=['label'])
train_Y = train_dataset['label']
```

```python
train_X[categorical_columns].dtypes
```

```python
# 檢查類別資料的不相同數量
train_X[categorical_columns].nunique()
```

```python
test_X[categorical_columns].nunique()
```

```python
dataset_X = pd.concat([train_X, test_X])

# 確定沒做/可能沒做 label encoding 的類別資料
not_encoded_columns = []
probably_not_encoded_columns = []

# 確認哪個類別資料還沒做label encoding
for column in categorical_columns:
    if dataset_X[column].dtype == 'object':
        not_encoded_columns.append(column)
    else:
        unique_values = dataset_X[column].nunique()

        (max_value, min_value) = (dataset_X[column].max(), dataset_X[column].min())
        expected_unique_values = int(max_value - min_value + 1)

        print(f'{column} (實際/預期): {unique_values}/{expected_unique_values}')
        if expected_unique_values == unique_values:
            pass
        else:
            probably_not_encoded_columns.append(column)


encoded_columns = list(set(categorical_columns) - set(not_encoded_columns) - set(probably_not_encoded_columns))
print(f'\n未做 label encoding 的類別資料: {not_encoded_columns}')
print(f'可能未做 label encoding 的類別資料: {probably_not_encoded_columns}')
print(f'已經做 label encoding 的類別資料: {encoded_columns}')
```

```python
dataset_X[probably_not_encoded_columns].describe()
```

```python
dataset_X[encoded_columns].describe()
```

```python
# 對類別資料做 label encoding
# 匯入 LabelEncoder
from sklearn.preprocessing import LabelEncoder

# 創建 LabelEncoder
all_need_to_encoded_columns = list(set(not_encoded_columns) | set(probably_not_encoded_columns))
label_encoders = {column: LabelEncoder() for column in all_need_to_encoded_columns}

# 對類別資料做 label encoding
for column, label_encoder in label_encoders.items():
    dataset_X[column] = label_encoder.fit_transform(dataset_X[column])
```

## 拆分為訓練測試集

```python
# 將資料集分回訓練集與測試集
train_X = dataset_X.iloc[:len(train_X)]
test_X = dataset_X.iloc[len(train_X):]
```

```python
# 驗證資料集的維度
print(train_X.shape)
print(train_Y.shape)
print(test_X.shape)
```

## 對數值類資料做正規化

```python
# 找出數值型態的欄位
numeric_columns = list(set(train_X.columns) - set(categorical_columns))
```

```python
train_X[numeric_columns].describe()
```

```python
test_X[numeric_columns].describe()
```

```python
from sklearn.preprocessing import StandardScaler

# 創建 StandardScaler
scaler = StandardScaler()

# 在訓練集上計算正規化參數 (numpy array)
train_X_scaled = scaler.fit_transform(train_X[numeric_columns])

# 使用訓練集的正規化參數對測試集進行正規化 (numpy array)
test_X_scaled = scaler.transform(test_X[numeric_columns])

# 將正規化後的資料轉換回 DataFrame
train_X.loc[:, numeric_columns] = train_X_scaled
test_X.loc[:, numeric_columns] = test_X_scaled
```
